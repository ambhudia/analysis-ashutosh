# Getting the SZZ from MOHID output

```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from salishsea_tools import utilities
%matplotlib inline
```

```python
path = '/home/abhudia/viz_test/Lagrangian_SOGa_01dec17_MF0_1000m3_7days_const_w_unrotated.hdf5'
f = h5py.File(path)
vertical_z = f['Grid']['VerticalZ']
keys = []
for key in vertical_z.keys():
    keys.append(key)

number_steps = len(keys)

output = np.zeros([number_steps, 41])
output_avg = np.zeros([number_steps, 41])
bar = utilities.statusbar('Loading...', maxval = number_steps)
for i, key in bar(enumerate(keys)):
    array = np.asarray(vertical_z[key])
    output[i] = np.max(np.max(array, axis = 1), axis = 1)
```

```python
fig = plt.figure(figsize = (10,10))
plt.subplot(1,2,1)
#plt.pcolormesh(output.T)
#plt.colorbar()
plt.plot(output)
plt.ylabel('Depth (m)')
plt.xlabel('Index of hour')
plt.title('All Layers')
plt.gca().invert_yaxis()
plt.subplot(1,2,2)
plt.plot(output[...,0:10])
plt.ylabel('Depth (m)')
plt.xlabel('Index of hour')
plt.title('Top 10 Layers')
plt.gca().invert_yaxis()
```
