# Compare the 1 hour interval currents on the full SSC grid vs the 20 mins interval currents over a part of the domain 
- Apply a low pass and band pass filter on the 20 mins current speeds at the three chosen locations to look at the fornightly cycles
- Apply a running average on the 20 min interval currents to reduce the noise in the signal

## Create Plots

```python
def average_output(data_array, times, period):

    unique, counts = np.unique(np.asarray(times).astype('datetime64[{}]'.format(period)).astype(datetime.datetime), return_counts=True) 
    index = 0; count_so_far = 0; means = []
    
    for count in counts:
        means.append(np.sum(data_array[count_so_far: count_so_far + count])/count)
        count_so_far = count_so_far + count
        
    return([unique, means])  
```

## Plots with low pass and band pass filters

```python
locs = ['sog', 'salmon', 'turn']
locations = ['Strait of Georgia', 'Salmon Bank', 'Turn Point']
years = [2015,2016, 2017, 2018] 

for index in range(3):
    for year in years:
        current = locals()['current_{}{}'.format(locs[index], str(year))]
        times = []
        for time in times_current: 
            if time.year == year:
                times.append(time)
        fig = plt.figure(figsize=(16,9))
        ax = plt.subplot(111)
        plt.plot(times, current, 'r-', label = 'raw')
        plt.xlabel('Time Axis')
        plt.ylabel('Current Speed (m/s)')
        title = 'Current speed at {} in {} (m/s)'.format(locations[index], year)
        
        # using a low pass and band pass filter
        order = 2
        fs = 1 / (60*20)
        lowcut = 1 / 86400 / 14
        highcut = 1 / 86400 /10
        nyq = 0.5 * fs
        low = lowcut / nyq
        high = highcut / nyq
        #low_pass = lfilter(*butter(order,low,btype = 'low'),current)
        #plt.plot(times, low_pass, 'b-', label = 'low pass filter')
        low_pass_filt = filtfilt(*butter(order,low,btype = 'low'),current)
        plt.plot(times, low_pass_filt, 'b-', label = 'low pass filter')
        band_pass = filtfilt(*butter(order,[low, high],btype = 'band'),current)
        plt.plot(times, band_pass, 'g-', label = 'band pass filter')
        ax.set_xlim(times[0], times[-1])
        
        ax.legend(prop={'size': 20})
        plt.title(title)
        for item in ([ax.xaxis.label, ax.yaxis.label] +
            ax.get_xticklabels() + ax.get_yticklabels()):
            item.set_fontsize(16)
        # set the font size for the axis label
        ax.title.set_fontsize(20)
```

### Plots with running averages

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
locs = ['sog', 'salmon', 'turn']
locations = ['Strait of Georgia', 'Salmon Bank', 'Turn Point']
years = [2015, 2016, 2017, 2018] 
N = 72
for index in range(3):
    for year in years:
        current = locals()['current_{}{}'.format(locs[index], str(year))]
        times = []
        for time in times_current: 
            if time.year == year:
                times.append(time)
        fig = plt.figure(figsize=(25,10))
        plt.plot(times, current, 'r-', label = 'raw')
        plt.xlabel('time')
        plt.ylabel('Current Speed (m/s)')
        title = 'Current speed at {} in {} (m/s)'.format(locations[index], year)
        avg_values = running_mean(current, N)
        avg_times = times[int(N/2): int(len(times) - N/2 +1)]
        plt.plot(avg_times, avg_values, 'b-', label = 'running average')
        plt.legend()
        plt.title(title)
```
